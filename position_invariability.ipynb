```python
import matplotlib.pyplot as plt
from PIL import Image, ImageOps

# Define the border width in pixels and the border color
BORDER_WIDTH = 10
BORDER_COLOR = 'black'

# Paths to your images
image_paths = [
    "data/Modhealthy/008_C_NUE_R1-10-06-2019 20_44_23-1043.jpg",
    "data/Modhealthy/041_C_NUE_R1-05-06-2019 16_12_59-1241.jpg",
    "data/Modhealthy/026_C_NUE_R2-05-06-2019 16_22_01-1152.jpg"
]

# Function to display images with a black border
def display_images_with_border(image_paths, border_width, border_color):
    plt.figure(figsize=(15, 10))
    
    for i, image_path in enumerate(image_paths):
        img = Image.open(image_path)
        bordered_img = ImageOps.expand(img, border=border_width, fill=border_color)
        
        plt.subplot(1, len(image_paths), i+1)
        plt.imshow(bordered_img)
        plt.axis('off')  # Turn off the axis

    plt.tight_layout()
    plt.show()

# Call the function to display the images
display_images_with_border(image_paths, BORDER_WIDTH, BORDER_COLOR)
```
